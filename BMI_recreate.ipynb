```python
def get_degrees(data):
    g = nx.from_pandas_adjacency(data,create_using=nx.DiGraph)
    in_degree = dict(g.in_degree)
    out_degree = dict(g.out_degree)
    total_degrees = {}
    for node in g.nodes():
        total_degrees[node] = in_degree[node] + out_degree[node]
    return in_degree, out_degree, total_degrees
```

```python
def degree_probability(x):
    counts = pd.Series(x).value_counts()
    prob = counts / counts.sum()
    return prob
# counts = pd.Series(in_degree).value_counts()
# prob = counts / counts.sum()
```

```python
def add_trend(data, color, label):
    filtered_data = data[(data.index > 0) & (data.values > 0)]
    if len(filtered_data) < 2:
        return label
    log_degree = np.log10(filtered_data.index)
    log_prob = np.log10(filtered_data.values)
    slope, intercept = np.polyfit(log_degree, log_prob, 1)
    power_y = 10 ** (intercept + slope * log_degree)
    plt.plot(filtered_data.index, power_y, color=color, linestyle='-')
    return f"{label} = {slope:.3f}"
```

```python
def plot_scatter(total_dg,out_dg,in_dg):
    plt.figure(figsize=(7, 5))
    sns.scatterplot(x=total_dg.index, y=total_dg.values, color='blue', label = 'total_degree')
    sns.scatterplot(x=out_dg.index, y=out_dg.values, color='red', label = 'out_degree')
    sns.scatterplot(x=in_dg.index, y=in_dg.values, color='green', label = 'in_degree')
    total_slope = add_trend(total_dg, 'blue', 'total') 
    in_slope = add_trend(in_dg, 'green', 'total') 
    out_slope = add_trend(out_dg, 'red', 'total') 

    plt.xscale('log')
    plt.yscale('log')
    plt.legend([total_slope,in_slope,out_slope])
    plt.xlabel('degree k')
    plt.ylabel('Pr(k)')
    plt.title(sheet_names[sheet_num])
    plt.show()
```

```python
sheet_names = ['E.coli (iJ01366)', 'B. subtilis (iYO844)', 'G. metallireducens (iAF987)', 'K. pneumoniae (iYL1228)', 'S. cerevisiae (iMM904)']
sheet_num = 0
raw_df = pd.read_excel('/Users/kevindai/UCSF_Fall_2025/gwas 2/12859_2019_2897_MOESM2_ESM.xlsx',
                       index_col=0, sheet_name=sheet_names[sheet_num])
g = nx.from_pandas_adjacency(raw_df,create_using=nx.DiGraph)
in_degree, out_degree, total_degrees = get_degrees(raw_df)
in_dg = degree_probability(in_degree)
out_dg = degree_probability(out_degree)
total_dg = degree_probability(total_degrees)
plot_scatter(total_dg,out_dg,in_dg)
```
